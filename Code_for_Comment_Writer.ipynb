```python
#Final Project Code:

def student_comment(file):
    import csv
    with open(f"data/{file}") as f:
        data = csv.reader(f, delimiter=',')
        header = next(data)
        for row in data:
            student_name = row[1]

            def overview(row:list):
                return f"{row[1]} {row[0]}:\n\nOverview: It has been a joy teaching you in {row[2]} this semester. We have covered a wide variety of topics in {row[2]} through many different means, such as homework, a midterm exam, and a final exam. You even conducted an independent research project on the {row[8]}! I hope to teach you again in the future!"

            overview(row)

            def comment1(row:list):
                if row[12] == "A":
                    grade = "amazing"
                if row[12] == "B":
                    grade = "good"
                if row[12] == "C":
                    grade = "ok"
                if row[12] == "D":
                    grade = "deally bad"
                if row[12] == "F":
                    grade = "horrible"

                if int(row[4]) > int(row[3]):
                    if int(row[4]) - int(row[3]) > 15:
                        improvement = "great improvement!"
                    if int(row[4])  - int(row[3]) > 0:
                        improvement = "improvement."
                else:
                    improvement = "worse performance than before you started. What are you even doing?"

                return f'Results: Overall, this year you did {grade}. On your midterm you recieved a {row[3]}, and on your final you recieved a {row[4]}. As such, you have shown {improvement} Your presentation on the {row[8]} was {row[9]}, with a score of {row[10]}. It is nice to see you begin to synthesize new connections, however.'

            comment1(row)

            def comment2(row:list):
                #collaboration indicator
                if row[15] == '4':
                    q1 = "Q1 Learning Indicator: Collaboration\nYou are an amazing collaborator. You are truly a gem to work with. You contribute to all class discussions and activities but never take the spotlight from others. You are constructive and provide effective feedback to others without just letting them copy your work. Indicator grade: Strong\n\n"
                elif row[15] == '3':
                    q1 = 'Q1 Learning Indicator: Collaboration\nYou are a solid collaborator. There are times where you may dominate a class discussion or not speak up enough, but overall you contribute to the class as expected. Indicator Grade: Proficient\n\n'
                elif row[15] == '2':
                    q1 = "Q1 Learning Indicator: Collaboration\nYou sometimes contribute to class discussions, but other times you refuse to participate. You also have a tendency to copy other people's work a little too close word for word. Indicator Grade: Approaching\n\n"
                elif row[15] == '1':
                    q1 = "Q1 Learning Indicator: Collaboration\nYou are a disgrace to humanity. You copy other people's work verbatim but do not share your work when your peers ask for it. You also yell at peers when they make a mistake and blame others when you make a mistake. You need accountability in your life right now. Indicator Grade: Beginning\n\n"

                #showing work indicator
                if row[16] == '4':
                    q2 = "Q2 Learning Indicator: Ability to Show Work\nYou always show your thinking through proofs, diagrams, and thinking. You are a model student that everyone should look up to. Indicator Grade: Strong\n\n"
                if row[16] == '3':
                    q2 = "Q2 Learning Indicator: Ability to Show Work\nYou usually show your thinking effectively, but there are times where you do some of the thinking in your head and do not express how you got there on the paper. Indicator Grade: Proficient\n\n"
                if row[16] == '2':
                    q2 = "Q2 Learning Indicator: Ability to Show Work\nShowing work is an area of growth for you. Although you usually get the correct answers, I have no idea how you got there. Indicator Grade: Approaching\n\n"
                if row[16] == '1':
                    q2 = "Q2 Learning Indicator: Ability to Show Work\nYou never show work or even respond to the problem at hand. It is clear that you do not care about math nor read the instructions of the problem. You need help. Indicator Grade: Beginning\n\n"

                #smartness indicator
                if row[17] == '4':
                    q3 = "Q3 Learning Indicator: Smartness\nYou are one of the smartest students I have ever seen. You are incredibly professional, and the way you interact with me, your peers, and your calculus work shows that you could very well be the next Einstein. Indicator Grade: Strong\n\n"
                if row[17] == '3':
                    q3 = "Q3 Learning Indicator: Smartness\nYou are a gifted student and of great inteligence. You are well on your way to succeeding to an elite university. Indicator Grade: Proficient\n\n"
                if row[17] == '2':
                    q3 = "Q3 Learning Indicator: Smartness\nThere are moments where I see your inteligence coming through, but sometimes I question your logical thinking. Occasionally it takes you longer to comprehend new subjects, but that is okay. I believe that you could thrive at a local community college before transferring to a more rigorous institution. Indicator Grade: Approaching\n\n"
                if row[17] == '1':
                    q3 = "Q3 Learning Indicator: Smartness\nYou show no signs of brain activity. I don't understand how it is possible to not be able to answer the questions you are unable to answer. I don't know how to help you. This is beyond me. I'm sorry. Indicator Grade: Beginning\n\n"

                #overall indicator
                if (int(row[15]) + int(row[16]) + int(row[17]))/3 > 3:
                    overall = "Indicator Wrap-Up: you show great promise as an academic with a blend of inteligence, hard-work, and the ability to collaborate. You are well-rounded student. You have the tools to succeed at anything you put your mind to."
                elif (int(row[15]) + int(row[16]) + int(row[17]))/3 > 2:
                    overall = "Indicator Wrap-Up: you ocassionally show glimpses of elite student work in my class. There are still areas for you to work on, but I see your potential."
                elif (int(row[15]) + int(row[16]) + int(row[17]))/3 > 1:
                    overall = "Indicator Wrap-Up: you struggle with many of the learning indicators. This is an area of growth for you but the hard work will be worth it. These tools are necessary to have longterm success."
                else:
                    overall = "Indicator Wrap-Up: you have scored the lowest possible scores on all of the indicators and do not show much potential in academics. Maybe summer school would be beneficial for your learning."

                #paragraph
                return f"{q1}{q2}{q3}{overall}"

            comment2(row)

            def conclusion(row:list):
                if row[12] == "A":
                    reaction = "amazed"
                elif row[12] == "B":
                    reaction = "happy"
                elif row[12] == "C":
                    reaction = "satisfied"
                elif row[12] == "D":
                    reaction = "disappointed"
                elif row[12] == "F":
                    reaction = "disgusted"

                if row[12] == "A":
                    final_grade = "you were one of the best students in the class, and I am happy to have you in my class."
                elif row[12] == "B":
                    final_grade = "you did hard work and did your best. I'm proud of you."
                elif row[12] == "C":
                    final_grade = "you were truly one of the most average students of all time."
                elif row[12] == "D":
                    final_grade = "you could certainly use some improvement."
                elif row[12] == "F":
                    final_grade = "get out of my classroom and never return."

                if int(row[13]) > 5:
                    attendance = f'Furthermore, because of your high number of absences in {row[2]}, you fail my class. How did you even accumulate {row[13]} absences in a semester? This is a disgrace. I hope to never see you in my class again!!!'
                else:
                    attendance = ''
                return (f'Conclusion: I was {reaction} at your performance in {row[2]}. With a final grade of {row[11]}%, or a letter grade of {row[12]}, {final_grade} {attendance} In short, you {row[14]}ED this class.')

            conclusion(row)


            one = overview(row)
            two = comment1(row)
            three = comment2(row)
            four = conclusion(row)
            final_comment = f"{one.strip()}\n\n {two.strip()}\n\n {three.strip()}\n\n {four.strip()}"
            with open(f'data/{student_name}.txt','w') as w:
                w.write(final_comment)
student_comment('teacher.csv')
```
